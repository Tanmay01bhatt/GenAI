LangGraph Basics

1 - State + Schema

```python
from typing_extensions import TypedDict
```

```python
class State(TypedDict):
    graph_state: str
```

2 - Node

```python
def node_1(state):
    print("---Node 1---")
    return {"graph_state": state['graph_state'] + " AGI"}

def node_2(state):
    print("---Node 2---")
    return {"graph_state": state['graph_state'] + " Achieved!"}

def node_3(state):
    print("---Node 3---")
    return {"graph_state": state['graph_state'] + " Not Achieved :("}
```

3 - Edge

```python
import random
from typing import Literal
```

```python
def decide_mood(state) -> Literal["node_2", "node_3"]:
    # Often, we use the state to decide the next node
    user_input = state['graph_state']
    
    # Here, we randomly split between Node 2 and Node 3
    if random.random() < 0.5:
        # 50% of the time, we return Node 2
        return "node_2"
    
    # Otherwise, we return Node 3
    return "node_3"
```

Create A Graph

State

```python
from langgraph.graph import StateGraph
builder = StateGraph(State)
```

Add the Nodes

```python
builder.add_node("node_1", node_1)
builder.add_node("node_2", node_2)
builder.add_node("node_3", node_3)
```

Mark Staring Node

```python
from langgraph.graph import START, END

builder.add_edge(START, "node_1")
```

Add the conditional edge

```python
builder.add_conditional_edges("node_1", decide_mood)
```

Connect the nodes

```python
builder.add_edge("node_2", END)
builder.add_edge("node_3", END)
```

Compile

```python
graph = builder.compile()
```

Output

```python
print(graph.invoke({"graph_state" : "Has AGI been achieved?"}))
```
